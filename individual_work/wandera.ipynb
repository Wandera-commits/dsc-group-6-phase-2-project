## Data Cleaning

```python
#Removing Duplicates
cleaned_df = df.drop_duplicates()
```

```python
#Handling missing values (example: filling with median)
df_without_missing_values = cleaned_df.fillna(cleaned_df.median(), inplace=False)
```

```python
df_without_missing_values.info()
```

## Data PreProcessing

```python
from datetime import datetime
```

```python
# Correcting data types (example: converting date column to datetime)
df_without_missing_values['sqft_basement'] = pd.to_numeric(df_without_missing_values['sqft_basement'], errors='coerce')
df_without_missing_values['date'] = pd.to_datetime(df_without_missing_values['date'])
```

```python
# Correlation matrix to determine columns to drop
corr_matrix = df_without_missing_values.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Drop unnecessary columns
df_with_dropped_columns = df_without_missing_values.drop(columns=['id', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'view', 'waterfront', 'condition', ])
```

```python
df_with_dropped_columns.info()
```

```python
# Correlation after dropped columns
corr_matrix = df_with_dropped_columns.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Removing Outliers
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_with_dropped_columns['bedrooms'].quantile(0.05)
Q3 = df_with_dropped_columns['bedrooms'].quantile(0.95)

# Calculate IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_no_outliers = df_with_dropped_columns[(df_with_dropped_columns['bedrooms'] >= lower_bound) & (df_with_dropped_columns['bedrooms'] <= upper_bound)]
```

```python
print("Original DataFrame:")
print(df_with_dropped_columns)
print("\nDataFrame without outliers:")
print(df_no_outliers)
```
