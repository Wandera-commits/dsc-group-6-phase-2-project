# Sort the null values
We check for missing values

```python
# Check for missing values

kcdata.isnull().sum()
```

We note that *waterfront* has 2376 missing values, *view* has 63 missing values and *yr_renovated* has 3842 missing values

```python
# Finding percentage of the missing values above

for col in ['waterfront', 'view', 'yr_renovated']:
    percent_missing = kcdata[col].isnull().sum() * 100 / len(kcdata)
    print(f"{col}:{round(percent_missing,3)}")
```

```python
# Decription of the missing values

kcdata[['waterfront', 'view', 'yr_renovated']].describe ()
```

In conclusion, we should replace the missing values with zero

```python
# Function to fill the missing values with zero

def fill_missing_values(kcdata):
    ''' Fill missing values with zero'''
    for col in ['waterfront', 'view', 'yr_renovated']:
        kcdata[col] = kcdata[col].fillna(0)
    return kcdata

# Application of the function
kcdata = fill_missing_values(kcdata)
```

```python
# Check if there are any more missing values

kcdata.isna().any().any()
```
